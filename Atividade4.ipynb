**Introdução a Guias de Onda Retengulares**

Considerando um guia de onda como o ilustrado na figura abaixo, cujas paredes de largura *a* e altura *b* são condutores perfeitos, preenchido por um meio dielétrico sem perdas e sem cargas.
imagem.png

*Figura 1: Imagem retirada do livro Elementos de Eletromagnetismo, Sadiku, 7º edição, cap.12.*

Recapitulando as Equações de Maxwell da forma fasorial para dielétricos sob essas condições, temos que:


 \begin{equation}
 \overrightarrow{\nabla}\cdot\overrightarrow{E_{s}} = 0
 \end{equation}

 \begin{equation}
 \overrightarrow{\nabla}\cdot\overrightarrow{H_{s}} = 0
 \end{equation}

 \begin{equation}
\overrightarrow{\nabla}\times\overrightarrow{E_{s}} = -j\omega \mu \overrightarrow{H_{s}}
 \end{equation}

\begin{equation}
\overrightarrow{\nabla}\times\overrightarrow{H_{s}} = j\omega \varepsilon  \overrightarrow{E_{s}}
\end{equation}
