**Estudo do Campo de Potencial Elétrico em um Guia de Onda Retangular**

Considerando um guia de onda como o ilustrado na figura abaixo, cujas paredes de largura *a* e altura *b* são condutores perfeitos, preenchido por um meio dielétrico sem perdas e sem cargas.

imagem.png

*Figura 1: Imagem retirada do livro Elementos de Eletromagnetismo, Sadiku, 7º edição, cap.12.*

Para o estudo do campo de potencial dentro do guia, deve-se lembrar da Lei de Gauss e da relação entre campo e potencial elétrico, dadas por:
\begin{equation}
 \vec{\nabla}\cdot\vec{E}=\frac{\rho _{v}}{\varepsilon }
 \\
 \vec{E}=-\nabla V
\end{equation}

Substituindo a eq. (II) na (I), para um meio sem cargas, tem-se que:

\
\begin{equation}
\nabla^{2}V=0
\end{equation}

Que, pela simetria do problema, se resume à equação de Laplace em duas dimensões:

\
\begin{equation}
\frac{\partial^2 V}{\partial x^2}+\frac{\partial^2 V}{\partial y^2} = 0
\end{equation}

Para a resolução da EDP acima, inicia-se assumindo a hipótese de que V(x,y) = X(x) Y(y):

\
\begin{equation}
\frac{\partial^2 (XY)}{\partial x^2}+\frac{\partial^2 (XY)}{\partial y^2} = 0
\end{equation}

Como Y(y) é constante em relação a x, e vice-versa:

\
\begin{equation}
\ Y \frac{\mathrm{d}^2 X}{\mathrm{d} x^2}+ X \frac{\mathrm{d}^2 Y}{\mathrm{d} y^2} = 0
\end{equation}

Dividindo ambos os lados da equação por X(x) Y(y):

\
\begin{equation}
\frac{X''}{X} + \frac{Y''}{Y} = 0
\end{equation}

\
\begin{equation}
\frac{X''}{X} = - \frac{Y''}{Y} = \lambda
\end{equation}

Em que cada um dos membros é igual a um número λ .

A equação parcial, portanto, se restringe a um sisetma de duas equações ordinárias:

\
\begin{matrix}
X'' & -\lambda X & = 0 \\
Y'' & +\lambda Y & = 0  \\
\end{matrix}

E é fácil notar que as soluções para X(x) e Y(y) serão dadas por:

\
\begin{matrix}
X & = A sinh(\sqrt{\lambda}x) & + B cosh(\sqrt{\lambda}x) \\
Y & = A' sin(\sqrt{\lambda}y) & + B' cos(\sqrt{\lambda}y)  \\
\end{matrix}

Em que os coeficientes A, B, A' e B' serão determinados pelas condições de fronteira do problema.

**Princípio da Sobreposição**

Para descrever o campo potencial dentro do guia, deve-se trabalhar com as quatro condições de fronteira abaixo, que correspondem aos potenciais elétricos nas parades condutoras:

\
\begin{matrix}
V(x, 0) & = V_{1} \\
V(a, y) & = V_{2} \\
V(x, b) & = V_{3} \\
V(0, y) & = V_{4} \\
\end{matrix}

Devido à complexidade da solução desse sistema, opera-se o cálculo considerando cada uma das condições não homogêneas individualmente, para então ser feita a soma das quatro soluções.

**Para a condição não homogênea em V(a,y)**

Nesse caso particular, admite-se as novas condições de fronteira:

\
\begin{matrix}
V(x, 0) & = 0 \\
V(a, y) & = V_{2} \\
V(x, b) & = 0 \\
V(0, y) & = 0 \\
\end{matrix}

Cujas equações homogêneas são substituídas nas soluções-geral de forma que:

\
\begin{equation}
\ X(0) = 0
\end{equation}
\
\begin{equation}
\ 0 = Asinh(0) + Bcosh(0)
\\
\ {\color{DarkRed} {B=0}}
\end{equation}

\
\begin{equation}
\ Y(0) = 0
\end{equation}
\
\begin{equation}
\ 0 = A'sin(0) + B'cos(0)
\
\ {\color{DarkRed} {B'=0}}
\end{equation}

\
\begin{equation}
\ Y(b) = 0
\end{equation}
\
\begin{equation}
\ 0 = A'sin(\sqrt{\lambda} b)
\end{equation}

E a única forma dessa igualdade ser verdadeira e Y(y) ser diferente de zero (e, portanto, X(x)Y(y) também o ser,) ocorre quando:

\
\begin{equation}
{\color{DarkRed}{\sqrt{\lambda} = \frac{n\pi}{b}}}
\end{equation}

Para n um número inteiro positivo.

O conjunto de condições homogêneas permite que uma gama de soluções com o formato abaixo satisfaça o sistema:

\
\begin{equation}
\ \ {\color{DarkRed} {V(x, y) = \sum_{n=1}^{\infty}c_{n}sinh(\frac{n\pi x}{b})sin(\frac{n\pi y}{b})}}
\end{equation}

Mas apenas um conjunto de coeficientes cn satisfaz a condição não homogênea de V(a,y). A fim de calculá-lo, obtém-se que:

\
\begin{equation}
\ V(a, y) = V_{2} = \sum_{n=1}^{\infty}c_{n}sinh(\frac{n\pi a}{b})sin(\frac{n\pi y}{b})
\end{equation}

O desafio para se resolver esse problema é extrair o coeficiente cn do somatório e obter uma expressão em função de n.

Para isso, deve-se lembrar da relação trigonométrica:

\begin{equation}
\ {\color{DarkBlue} {sin(a)\cdot sin(b) = \frac{1}{2}\left [ cos(a-b)-cos(a+b) \right ]}}
\end{equation}

E também de que a integral de meio período da função cosseno é nula, a menos que o argumento seja zero:

\
\begin{equation}
\ {\color{DarkBlue} {\int_{0}^{b} cos( \frac{n \pi x}{b} ) \cdot dx = \begin{Bmatrix}
0, & n \neq 0\\
b & n =0
\end{Bmatrix}}}
\end{equation}

Recapitulando essas relações, nota-se que num intervalo de meio período fundamental, dois senos são ortogonais se, e somente se, seus argumentos forem diferentes.

Aplicando essa definição:

\
\begin{equation}
\ V_{2} \int_{0}^{b} sin(\frac{m\pi y}{b})\cdot dy= (\sum_{n=1}^{\infty}c_{n}sinh(\frac{n\pi a}{b})sin(\frac{n\pi y}{b})) \int_{0}^{b} sin(\frac{m\pi y}{b})\cdot dy
\end{equation}

Uma vez que a integral não depende de n, pode-se incorporá-la ao somatório e aplicar o princípio da ortogonalidade dos senos:

\
\begin{equation}
\ V_{2} \int_{0}^{b} sin(\frac{m\pi y}{b})\cdot dy= \sum_{n=1}^{\infty}c_{n}sinh(\frac{n\pi a}{b}) \int_{0}^{b} sin(\frac{m\pi y}{b})sin(\frac{n\pi y}{b})\cdot dy
\end{equation}

\
\begin{equation}
\ V_{2} \cdot \frac{b}{m\pi} \left[ -cos(\frac{m\pi y}{b}) \right]_0^b = c_{m}sinh(\frac{m\pi a}{b}) \cdot \frac{b}{2}
\end{equation}

Pois a única parcela não nula do somatório ocorre quando n = m.

Observa-se que o primeiro membro da igualdade oscila para valores pares e ímpares do índice, de tal forma que encontramos o seguinte conjunto de coeficientes cn:

\
\begin{equation}
{\color{DarkRed}{
c_n = \begin{Bmatrix}
0, & n = 2, 4, 6, 8...\\
\frac{4 V_2}{n \pi} \cdot \frac{1}{sinh(n\pi a/b)} & n = 1, 3, 5, 7...
\end{Bmatrix}}}
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt

a = 3 # Largura do guia
b = 2 # Altura do guia
V2 = 20 # Potencial em V(a,y)

# Criando o domínio
x = np.linspace(0, a, 100)
y = np.linspace(0, b, 100)
X, Y = np.meshgrid(x, y)

def V(x,y):
  V = 0
  for n in range(1, 150, 2):
    num = 4 * V2 * np.sinh(n * np.pi * x / b) * np.sin(n * np.pi * y / b)
    den = n * np.pi * np.sinh(n * np.pi * a / b)
    V += num/den

  return V

V_grid = V(X,Y)

plt.contourf(X, Y, V_grid, levels=50, cmap="cividis")
plt.colorbar(label="Valor de potencial")
plt.title("Potencial Elétrico no guia de onda")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

**Para a condição não homogênea em V(0, y)**

Novamente, admitindo as novas condições de fronteira:

\
\begin{matrix}
V(x, 0) & = 0 \\
V(a, y) & = 0 \\
V(x, b) & = 0 \\
V(0, y) & = V_4 \\
\end{matrix}

E iniciando o cálculo pelas condições homogêneas:

\begin{equation}
\ Y(0)=0
\\{\color{DarkRed} {B'=0}}
\end{equation}
\
\begin{equation}
\ Y(b)=0
\\0= A'sin(\sqrt{\lambda} b)
\\{\color{DarkRed} { \sqrt{\lambda}= \frac{n\pi}{b}}}
\end{equation}
\
\begin{equation}
\ X(a)=0
\\0= A \cdot sinh(\frac{n\pi}{b}a) + B \cdot cosh(\frac{n\pi}{b}a)
\end{equation}

\
Dividindo ambos os membros por B cosseno hiperbólico:

\
\begin{equation}
\ 0= \frac{A}{B} \cdot tgh(\frac{n\pi}{b}a) + 1
\\{\color{DarkRed} {B=-A \cdot tgh(\frac{n\pi}{b}a)}}
\end{equation}

Sob essas condições, obtém-se as expressões para X(x) e Y(y), dadas por:

\
\begin{equation}
\\{\color{DarkRed} {Y(y) = A'\cdot sin(\frac{n\pi}{b}y)}}
\
\end{equation}

\
\begin{equation}
\ X(x) = A\cdot sinh(\frac{n\pi}{b}x) - A\cdot tgh(\frac{n\pi}{b}a) \cdot cosh(\frac{n\pi}{b}x)
\\
\end{equation}


Observa-se, no entanto, que esta última equação pode ser simplificada pela seguinte relação de trigonometria hiperbólica:

\
\begin{equation}
\ {\color{DarkBlue} {sinh(x+y)=sinh(x)\cdot cosh(y) + cosh(x) \cdot sinh(y)}}
\end{equation}
\
\begin{equation}
\ {\color{DarkBlue} {sinh(-x) = -sinh(x)}}
\end{equation}
\
\begin{equation}
\ {\color{DarkBlue} {cosh(-x) = cosh(x)}}
\end{equation}

De tal modo que se obtém:

\
\begin{equation}
\ X(x) = \frac{-A\cdot sinh(\frac{-n\pi}{b}x)\cdot cosh(\frac{n\pi}{b}a) - A\cdot sinh(\frac{n\pi}{b}a) \cdot cosh(\frac{-n\pi}{b}x)}{cosh(\frac{n\pi}{b}a)}
\
\end{equation}

\
\begin{equation}
\ {\color{DarkRed} {X(x) = \frac{-A\cdot sinh(\frac{n\pi}{b}\cdot (a-x))}{cosh(\frac{n\pi}{b}a)}}}
\\
\end{equation}

Finalmente, encontra-se a solução-geral dessa EDP:

\
\begin{equation}
\ {\color{DarkRed} {V(x,y)=\sum_{n=1}^{\infty}c_{n} \cdot \frac{sin(\frac{n\pi}{b}y) \cdot sinh(\frac{n\pi}{b}\cdot (a-x))}{cosh(\frac{n\pi}{b}a)}}}
\
\end{equation}

Novamente, as condições homogêneas garantem uma família de soluções que satisfazem o sistema. Mas só um conjunto de coeficientes cn satisfaz a condição de fronteira não homogênea:

\
\begin{equation}
\ V(0,y)= V_{4} = \sum_{n=1}^{\infty}c_{n} \cdot tgh(\frac{n\pi}{b}a) \cdot sin(\frac{n\pi}{b}y)
\
\end{equation}

E, de novo, pelo princípio de ortogonalidade entre dois senos de argumentos distintos, pode-se tirar o coeficiente cn do somatório:

\
\begin{equation}
\ V_{4} \cdot \int_0^b sin(\frac{m\pi}{b}y) dy= \sum_{n=1}^{\infty}c_{n} \cdot tgh(\frac{n\pi}{b}a) \cdot sin(\frac{n\pi}{b}y) \int_0^b sin(\frac{m\pi}{b}y) dy
\
\end{equation}

\
\begin{equation}
\frac{V_4 b}{m\pi}\cdot\left[ -cos(\frac{m\pi}{b}y) \right]_0^b = c_m tgh(\frac{m\pi}{b}a)\cdot \frac{b}{2}
\end{equation}

\
\begin{equation}
{\color{DarkRed}{
c_n = \begin{Bmatrix}
0, & n = 2, 4, 6, 8...\\
\frac{4 V_4}{n \pi} \cdot \frac{1}{tgh(n\pi a/b)} & n = 1, 3, 5, 7...
\end{Bmatrix}}}
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt

a = 3 # Largura do guia
b = 2 # Altura do guia
V4 = 10 # Potencial em V(0,y)

# Criando o domínio
x = np.linspace(0, a, 100)
y = np.linspace(0, b, 100)
X, Y = np.meshgrid(x, y)

def V(x,y):
  V = 0
  for n in range(1, 150, 2):
    num = 4 * V4 * np.sin(n * np.pi * y / b) * np.sinh(n * np.pi * (a-x) / b)
    den = n * np.pi * np.sinh(n * np.pi * a / b)
    V += num/den

  return V

V_grid = V(X,Y)

plt.contourf(X, Y, V_grid, levels=50, cmap="cividis")
plt.colorbar(label="Valor de potencial")
plt.title("Potencial Elétrico no guia de onda")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

**Para a condição não homogênea em V(x, b)**

\
\begin{matrix}
V(x, 0) & = 0 \\
V(a, y) & = 0 \\
V(x, b) & = V_3 \\
V(0, y) & = 0 \\
\end{matrix}
\
Devido à simetria do problema, para obtenção da solução desse sistema aproveita-se a equação obtida para a condição V(a, y) não homogênea sob as seguintes manipulações das variáveis independentes:

\\
\begin{equation}
\ x \leftarrow y
\
\\ y \leftarrow x
\end{equation}

\
Além disso, de forma análoga ao caso anterior,

\begin{equation}
\ X(a) = 0
\
\\ 0 = A\cdot sin(\sqrt{\lambda} a)
\end{equation}
\
\begin{equation}
\ \sqrt{\lambda} = \frac{n\pi}{a}
\end{equation}

\
\begin{equation}
\ V(x, b) = V_{4} = \sum_{n=1}^{\infty}c_{n}sinh(\frac{n\pi b}{a})sin(\frac{n\pi x}{a})
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt

a = 3 # Largura do guia
b = 2 # Altura do guia
V3 = 20 # Potencial em V(x,b)

# Criando o domínio
x = np.linspace(0, a, 100)
y = np.linspace(0, b, 100)
X, Y = np.meshgrid(x, y)

def V(x,y):
  V = 0
  for n in range(1, 150, 2):
    num = 4 * V3 * np.sinh(n * np.pi * y / a) * np.sin(n * np.pi * x / a)
    den = n * np.pi * np.sinh(n * np.pi * b / a)
    V += num/den

  return V

V_grid = V(X,Y)

plt.contourf(X, Y, V_grid, levels=50, cmap="cividis")
plt.colorbar(label="Valor de potencial")
plt.title("Potencial Elétrico no guia de onda")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

**Para a condição não homogênea em V(x, 0)**

\
\begin{matrix}
V(x, 0) & = V_1 \\
V(a, y) & = 0 \\
V(x, b) & = 0 \\
V(0, y) & = 0 \\
\end{matrix}

\
Dessa vez, manipula-se as variáveis independentes na solução onde a condição não homogênea ocorre em V(0,y), da mesma forma do item anterior.

```python
import numpy as np
import matplotlib.pyplot as plt

a = 3 # Largura do guia
b = 2 # Altura do guia
V1 = 5 # Potencial em V(x,0)

# Criando o domínio
x = np.linspace(0, a, 100)
y = np.linspace(0, b, 100)
X, Y = np.meshgrid(x, y)

def V(x,y):
  V = 0
  for n in range(1, 150, 2):
    num = 4 * V1 * np.sin(n * np.pi * x / a) * np.sinh(n * np.pi * (b-y) / a)
    den = n * np.pi * np.sinh(n * np.pi * b / a)
    V += num/den

  return V

V_grid = V(X,Y)

plt.contourf(X, Y, V_grid, levels=50, cmap="cividis")
plt.colorbar(label="Valor de potencial")
plt.title("Potencial Elétrico no guia de onda")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

**Solução do Caso Geral**

Quando todas as paredes do guia tiverem potenciais diferentes de zero, a distribuição de potencial no dielétrico será a soma dos quatro casos estudados anteriormente.

```python
import numpy as np
import matplotlib.pyplot as plt

a = 3 # Largura do guia
b = 2 # Altura do guia
V1 = 13 # Potencial em V(x,0)
V2 = 5 # Potencial em V(a,y)
V3 = 2 # Potencial em V(x,b)
V4 = 7 # Potencial em V(0,y)

# Criando o domínio
x = np.linspace(0, a, 100)
y = np.linspace(0, b, 100)
X, Y = np.meshgrid(x, y)

def V(x,y):
  V = 0
  for n in range(1, 150, 2):
    # Expressão numérica para as paredes verticais
    Va = (4*np.sin(n*np.pi*y/b)*(V2*np.sinh(n*np.pi*x/b)+ V4*np.sinh(n*np.pi*(a-x)/b))) / (n*np.pi*np.sinh(n*np.pi*a/b))
    Vb = (4*np.sin(n*np.pi*x/a)*(V1*np.sinh(n*np.pi*(b-y)/a) + V3*np.sinh(n*np.pi*y/a))) / (n*np.pi*np.sinh(n*np.pi*b/a))
    V = V + Va + Vb
  return V

V_grid = V(X,Y)

plt.contourf(X, Y, V_grid, levels=50, cmap="cividis")
plt.colorbar(label="Valor de potencial")
plt.title("Potencial Elétrico no guia de onda")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
